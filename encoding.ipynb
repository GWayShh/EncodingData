```python
import pandas as pd
import numpy as np
```

```python
data = {
    'Category': ['Low', 'Medium', 'High', 'Medium', 'Low', 'High'],
    'Color': ['Red', 'Blue', 'Green', 'Blue', 'Red', 'Green'],
    'Value': [10, 20, 30, 20, 10, 30]
}

df = pd.DataFrame(data)

df
```

```python
# ordinal encoding
map = {'Low': 1, 'Medium': 2, 'High': 3}  # setting the order myself
categoryList = []
for category in df['Category']:
    categoryList.append(map[category])

df['Category_Ordinal'] = categoryList

# one hot encoding
uniqueColors = df['Color'].unique()  # get all the unique colors in the field
# making matrix of zeroes with same number of rows as in df and columns equal to number of unique colors
oneHotEncoding = np.zeros((df.shape[0], len(uniqueColors))) 

for i, color in enumerate(uniqueColors):
    oneHotEncoding[:, i] = (df['Color'] == color).astype(int)  # setting the ith column as a column of 0 and 1 according to if the color is present or not

oneHotdf = pd.DataFrame(oneHotEncoding, columns=[f'Color_{color}' for color in uniqueColors])  # making df with column names
df = pd.concat([df, oneHotdf], axis=1)  # concatenating it to original df

df
```
